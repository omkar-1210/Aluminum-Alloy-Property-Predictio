```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import mean_absolute_percentage_error, r2_score
from scikeras.wrappers import KerasRegressor
from skopt import BayesSearchCV
import lime
import lime.lime_tabular
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
file_path = r"C:\Users\loneo\OneDrive\Documents\mini project\al data.xlsx"
data = pd.read_excel(file_path)

# Separate features and target variables
X = data.drop(columns=['Sr.no', 'YS (MPa)', 'TS (MPa)', 'EL (%)'])  # Features (compositions and processing parameters)
y = data[['YS (MPa)', 'TS (MPa)', 'EL (%)']]  # Target (mechanical properties)

# Split the data into training and testing sets (9:1 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Define the DNN model creation function
def create_model(n_layers=3, n_units=64, activation='relu', learning_rate=0.001):
    model = models.Sequential()
    model.add(layers.InputLayer(input_shape=(X_train.shape[1],)))

    # Adding hidden layers with batch normalization
    for _ in range(n_layers):
        model.add(layers.Dense(n_units, activation=activation))
        model.add(layers.BatchNormalization())

    model.add(layers.Dense(3))  # 3 output nodes (YS, TS, EL)

    # Compile the model
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mean_squared_error')

    return model

# Wrapper for KerasRegressor using SciKeras
model = KerasRegressor(model=create_model, verbose=0)

# Define the hyperparameter search space for Bayesian optimization
param_distributions = {
    'model__n_layers': (1, 5),
    'model__n_units': (16, 128),
    'model__learning_rate': (1e-4, 1e-2, 'log-uniform'),
    'batch_size': (16, 64),
    'epochs': (50, 200),
    'model__activation': ['relu', 'tanh', 'elu']
}

# Bayesian optimization
bayes_search = BayesSearchCV(estimator=model, search_spaces=param_distributions, n_iter=200, cv=KFold(10), n_jobs=-1, verbose=1,random_state=42)
bayes_search.fit(X_train, y_train)

# Evaluate the model on the test data
y_pred = bayes_search.best_estimator_.predict(X_test)
mape = mean_absolute_percentage_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print model performance
print("Test MAPE: ", mape)
print("Test RÂ² score: ", r2)

# LIME setup for interpretability
explainer = lime.lime_tabular.LimeTabularExplainer(
    training_data=np.array(X_train),
    feature_names=X_train.columns,
    class_names=['YS (MPa)', 'TS (MPa)', 'EL (%)'],
    mode='regression'
)

# Function to plot LIME explanations
def plot_lime_explanation(exp, title):
    fig = exp.as_pyplot_figure()
    plt.title(title)
    plt.show()

# Predict and retrieve the top 5 recommendations
y_train_pred = bayes_search.best_estimator_.predict(X_train)
train_data = X_train.copy()
train_data['YS (MPa)'] = y_train_pred[:, 0]
train_data['TS (MPa)'] = y_train_pred[:, 1]
train_data['EL (%)'] = y_train_pred[:, 2]

# Calculate the product of YS and EL for sorting
train_data['YS*EL'] = train_data['YS (MPa)'] * train_data['EL (%)']
top_5 = train_data.sort_values(by='YS*EL', ascending=False).head(5)

# Output the top 5 recommended values with their composition, processing parameters, YS, TS, EL
print("Top 5 Recommendations:\n", top_5[['Zn', 'Mg', 'Cu', 'Zr', 'Ti', 'Fe', 'Si',
                                        'Casting     (0: B, 1:PM)',
                                        'Homogenization (celcius/h)',
                                        'Extrusion (0: X, 1:O)',
                                        'Solution Treatment (celcius/h)',
                                        'Aging (celcius/h)',
                                        'YS (MPa)', 'TS (MPa)', 'EL (%)']])

# Loop through the top 5 recommended compositions and interpret tensile properties
for index, row in top_5.iterrows():
    print(f"\nInterpretation for Recommended Composition #{index + 1}")
    
    # Fetch the recommended composition row
    sample = row[X.columns]  # Select only the feature columns
    
    # Generate explanations for each property
    exp_ys = explainer.explain_instance(sample, lambda x: bayes_search.best_estimator_.predict(x)[:, 0], num_features=5)
    exp_ts = explainer.explain_instance(sample, lambda x: bayes_search.best_estimator_.predict(x)[:, 1], num_features=5)
    exp_el = explainer.explain_instance(sample, lambda x: bayes_search.best_estimator_.predict(x)[:, 2], num_features=5)

    # Display interpretations for Yield Strength (YS) with visual explanation
    print("\nYield Strength (YS) Interpretation:")
    for feature, weight in exp_ys.as_list():
        print(f"{feature}: {weight:.3f}")
    plot_lime_explanation(exp_ys, "LIME Explanation for Yield Strength (YS)")

    # Display interpretations for Tensile Strength (TS) with visual explanation
    print("\nTensile Strength (TS) Interpretation:")
    for feature, weight in exp_ts.as_list():
        print(f"{feature}: {weight:.3f}")
    plot_lime_explanation(exp_ts, "LIME Explanation for Tensile Strength (TS)")

    # Display interpretations for Elongation (EL) with visual explanation
    print("\nElongation (EL) Interpretation:")
    for feature, weight in exp_el.as_list():
        print(f"{feature}: {weight:.3f}")
    plot_lime_explanation(exp_el, "LIME Explanation for Elongation (EL)")
```
